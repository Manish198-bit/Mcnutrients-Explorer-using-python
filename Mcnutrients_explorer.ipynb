```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mcd.csv',encoding= 'unicode_escape')
```

```python
df.info()
```

```python
**ENERGY DISTRIBUTION**
```

```python
sns.kdeplot(df['energy'], fill=True, color='teal')

plt.title('Energy Distribution (KDE)')
plt.xlabel('Energy (kcal)')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
** fat and sugar analysis **
```

```python
top_added = df.sort_values('added_sugar', ascending=False)

plt.barh(top_added['name'], top_added['added_sugar'], color='orchid')
plt.xlabel('Added Sugar (g)')
plt.title('Added Sugar by Food Item')
plt.show()
```

```python
PROTEIN VS ENERGY
```

```python
df_sorted = df.sort_values('protein')
plt.plot(df_sorted['protein'], df_sorted['energy'], marker='o', linestyle='-', color='blue')

plt.xlabel('Protein (g)')
plt.ylabel('Energy (kcal)')
plt.title('Protein vs Energy (Line Plot)')
plt.grid(True)
plt.show()
```

```python
ALLERGY TAG DISTRIBUTION
```

```python
df['allergy'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Allergy Tag Frequency')
plt.ylabel('Count')
plt.show()
```
